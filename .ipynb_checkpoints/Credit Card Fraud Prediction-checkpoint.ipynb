```python
plt.figure(figsize=(8,6))
colours = ['green', 'red']
sns.countplot(x = 'Class', data = df, palette= colours, hue = 'Class')
plt.title('Distribution of Fraud vs Normal Transaction \n (0 : Normal || 1 : Fraud)')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```
